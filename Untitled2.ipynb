🧩 Problem: Rotate Image (In-Place)

You are given an n x n 2D matrix representing an image, where each pixel is an integer. Rotate the image 90 degrees clockwise.

You must rotate the image in-place, meaning you cannot allocate another 2D matrix to perform the rotation.

⸻

Example:

Input:

```
matrix = [
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
]
```

Output (after rotation):
```
[
  [7, 4, 1],
  [8, 5, 2],
  [9, 6, 3]
]
```

⸻

Constraints:
* `matrix.length == n`
* `matrix[i].length == n`
* `1 <= n <= 1000`
* `Integers are arbitrary`

⸻

You’re looking to do two things:
	1.	Write a function that performs the rotation in-place.
	2.	Keep it clean and efficient.

Let me know if you want hints or edge cases to consider.
