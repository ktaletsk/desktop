```python
import numpy as np
import matplotlib.pyplot as plt

print(f"numpy version: {np.__version__}")
print("Ready to go!")
```

```python
# Generate a fun spiral plot
theta = np.linspace(0, 6 * np.pi, 1000)
r = theta**0.5

x = r * np.cos(theta)
y = r * np.sin(theta)

fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(x, y, c=theta, cmap="viridis", s=2)
ax.set_aspect("equal")
ax.set_title("Spiral")
ax.axis("off")
plt.tight_layout()
plt.show()
```
