```python
import matplotlib.pyplot as plt
import numpy as np

# Draw a smiley face
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect("equal")
ax.axis("off")
ax.set_facecolor("#FFF9C4")
fig.patch.set_facecolor("#FFF9C4")

# Face
face = plt.Circle((0, 0), 1, color="#FFD54F", ec="#F57F17", lw=3)
ax.add_patch(face)

# Eyes
left_eye = plt.Circle((-0.35, 0.3), 0.12, color="#4E342E")
right_eye = plt.Circle((0.35, 0.3), 0.12, color="#4E342E")
ax.add_patch(left_eye)
ax.add_patch(right_eye)

# Eye sparkles
left_sparkle = plt.Circle((-0.32, 0.34), 0.04, color="white")
right_sparkle = plt.Circle((0.38, 0.34), 0.04, color="white")
ax.add_patch(left_sparkle)
ax.add_patch(right_sparkle)

# Smile
theta = np.linspace(-np.pi * 0.8, -np.pi * 0.2, 100)
smile_x = 0.55 * np.cos(theta)
smile_y = 0.55 * np.sin(theta) + 0.05
ax.plot(smile_x, smile_y, color="#4E342E", lw=4, solid_capstyle="round")

# Rosy cheeks
left_cheek = plt.Circle((-0.6, -0.1), 0.12, color="#FF8A80", alpha=0.5)
right_cheek = plt.Circle((0.6, -0.1), 0.12, color="#FF8A80", alpha=0.5)
ax.add_patch(left_cheek)
ax.add_patch(right_cheek)

ax.set_title("Have a happy day! 😊", fontsize=18, fontweight="bold", pad=15)
plt.tight_layout()
plt.show()
```
