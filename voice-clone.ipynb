```
from mlx_audio.tts.utils import load_model

model = load_model("mlx-community/Kokoro-82M-bf16")
```

```
import numpy as np
from mlx_audio.tts.audio_player import AudioPlayer

player = AudioPlayer(sample_rate=24_000)
```

```
for res in model.generate("Clone test successful", voice="af_heart"):
  player.queue_audio(np.asarray(res.audio, dtype=np.float32).reshape(-1))
  player.wait_for_drain()


# player.queue_audio(np.asarray(res.audio, dtype=np.float32).reshape(-1))
# player.wait_for_drain()
# player.stop()  # drains then stops the stream
```
