# I'm Making My Ideal Notebook

This is not a post about the `.ipynb` file format by itself. It is a post about the application you write notebooks in, the runtime model behind it, and the kind of workflow notebooks should make possible.

A lot of the notebook world is fragmenting. Some people are still in JupyterLab. Others are moving to Marimo or Quarto. That shift is interesting, but the deeper question is simpler: what should the ideal notebook actually do for people?

## Thesis

The ideal notebook should preserve computation, not just text. It should let you close the window while work continues. It should keep outputs durable. It should be easy to jump into, easy to share, and built for a world where agents are not outsiders but first-class collaborators.

That means the notebook cannot be treated as a fragile front-end artifact. The kernel, the UI, and the coordinating server need to be decoupled. The notebook needs a real shared state model underneath it.

## Why Now

What pushed me here was not abstract collaboration theory. It was the very practical frustration of watching useful notebook state disappear. A long-running query finishes, a computation produces outputs that matter, and then the tab closes or the session dies and the visible result is gone.

That is absurd. People should be able to kick off substantial work, walk away, pick up their kid from school, come back, and still have the result. The ideal notebook should support life instead of forcing people to babysit a browser tab.

## The Manifesto

Here are the capabilities I want from an ideal notebook application.

1. **Durable execution**. If work is running, I should be able to close the window and return later without losing outputs.
2. **Agent-native collaboration**. Agents should be able to read cells, write cells, inspect outputs, and participate in the same state model as the human user.
3. **Instant entry**. It should feel as easy to start as opening an IPython shell, not like booting a small operating system.
4. **Real sharing**. A notebook should carry or discover enough environment information that another person can actually run it.
5. **Decoupled architecture**. The kernel, document state, and user interface should not collapse into a single fragile process boundary.
6. **Good defaults**. Settings, environments, and common workflows should work out of the box, but still remain inspectable and hackable.

## 1. Durable Execution

People call this background execution, but the kernel was always on the server side in many notebook systems. The real problem is that the notebook's visible state often is not. Outputs, execution progress, and the shape of work-in-flight too often depend on a live frontend session.

The ideal notebook fixes that gap. The computation should continue. The document should continue to reflect what happened. The result should still be there when I reconnect.

This is the feature that matters most to me. Not because it sounds futuristic, but because it solves the oldest and dumbest notebook pain: losing the thing you just spent time computing.

## 2. Agents Should Be Peers

Coding agents already work well in REPL-style loops because they can try something, inspect the result, and iterate. What they are missing most of the time is durable shared state. They set a variable, then have to fall back to writing whole scripts because the interaction model forgets too much between steps.

A notebook is a better substrate. It can preserve state across cells, preserve outputs, and give both humans and agents a shared artifact to work on. The ideal notebook should let an agent stream text into a cell, observe outputs, and build on prior computation without awkward handoffs.

That is not a hypothetical requirement. It should be a baseline expectation for notebook tools now.

## 3. Instant Entry

I do not want the notebook experience to feel heavy. Sometimes I just want something with the same ease as opening an IPython shell, trying an idea, and moving on. The ideal notebook should preserve that feeling of immediacy even while it supports richer execution, sharing, and collaboration.

This matters because exploratory work is fragile. Every bit of startup friction pushes people back toward less capable tools or toward ad hoc one-off scripts that are harder to revisit later.

## 4. Real Sharing

A notebook is only as useful as its ability to travel. If I hand it to someone else, they should be able to run it without guessing at packages, kernels, or environment setup.

The ideal notebook either carries its dependencies with it or points clearly at a project environment that can be recreated. It should make the trust and security story visible instead of magical, but it should also remove the miserable archaeology that usually follows when someone opens an old notebook and has no idea what it needs.

## 5. A Better Architecture

The kernel, the UI, and the coordinating server should be decoupled. The interface should be replaceable. The runtime should be durable. The document model should survive reconnects and support more than one kind of client.

That decoupling is what makes the other features possible. Without it, durable execution is brittle, agent collaboration is bolted on, and multi-window or multi-process workflows become accidental instead of fundamental.

## What nteract Desktop Already Demonstrates

Part of why I keep coming back to this problem is that pieces of the ideal notebook are already showing up in `nteract desktop`.

- It has a background daemon, `runtimed`, that handles notebook sync, kernel execution, and environment management.
- It supports realtime sync across windows and agents rather than treating the app as a single isolated editor.
- It uses the same Automerge model in the frontend and daemon, which makes shared document state feel like a real system primitive.
- It exposes Python bindings and an MCP server, which means notebooks are accessible to both code and agents, not just the GUI.
- It supports multi-client scenarios where two sessions share the same notebook document and kernel.
- It has a practical sharing story: notebooks remain standard `.ipynb` files, while dependencies can either travel inline or be discovered from project files.
- It syncs settings across windows using Automerge, which hints at the larger architectural direction: the app is centered on shared state, not one-off local UI state.

That does not mean the ideal notebook is finished. It means the future is starting to become concrete.

## Closing

So that is the manifesto: keep computation alive, keep outputs durable, make the notebook easy to enter, make it truly shareable, and treat agents as real participants.

If the notebook is going to remain one of the most important tools for thinking with code, it should stop pretending that the frontend session is the center of the universe. The document, the computation, and the collaboration model all deserve to be first-class.
