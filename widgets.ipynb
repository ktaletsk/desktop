```
import ipywidgets as widgets
from IPython.display import display
```

```
w_checkbox = widgets.Checkbox(value=False, description="Toggle me")
```

```
w_progress = widgets.IntProgress(
    value=0, min=0, max=10, description="Progress:", bar_style="info"
)
```

```
w_button = widgets.Button(description="Click Me!", button_style="info")
```

```
def on_click(b):
    w_checkbox.value = not w_checkbox.value
    w_progress.value = (w_progress.value + 1) % (w_progress.max + 1)
    w_progress.bar_style = "success" if w_progress.value == w_progress.max else "info"


w_button.on_click(on_click)

# %%
display(widgets.VBox([w_button, w_checkbox, w_progress]))
```
