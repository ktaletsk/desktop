# Building My Ideal Notebook

I want to build my ideal notebook — not the `.ipynb` file itself, but the **application you write notebooks in**.

People have been moving to [Marimo](https://marimo.io/), and that's an interesting shift worth exploring.

Some people are still using JupyterLab, others have moved to Marimo or Quarto — the ecosystem is fragmented.

So here's my plan: **list out all the killer features** I want in my ideal notebook app.

What's wild is that **an agent is writing this notebook right now**, just from me talking out loud. I'm dictating my thoughts and they're being structured into cells on the page in real time.

So let's get into it — here are the killer features I want in my ideal notebook app:

## Feature 1: Background Execution

I want to be able to **close the notebook window while cells are still running**. Computation should continue on the server, and I should be able to come back later and see the results.

The core problem is that Jupyter stores state in the **front end**, not server-side. If you close the browser, your outputs are gone. The ideal notebook keeps all execution state on the server — close the tab, reopen it, and everything is right where you left it.

It's not even really "background execution" — the kernel has *always* run on the server. The maddening part is that the **notebook file** doesn't persist there. The runtime is server-side, but your outputs aren't. That's the gap to fix.

I could build this as a JupyterLab plugin, but honestly I'd throw away almost everything and start fresh. The core experience I want is a dedicated space for **exploratory data analysis** — built from the ground up for that workflow.

Millions of people use notebooks — but many feel *trapped* in the current tooling. The goal isn't just to build something new. It's to **meet people where they are** and help them migrate forward. Move the standards forward, not just the product.
