```python
import math

# Generate a little multiplication table using list comprehension
size = 6
table = [[i * j for j in range(1, size + 1)] for i in range(1, size + 1)]

# Pretty print it
header = "    " + "".join(f"{j:>5}" for j in range(1, size + 1))
print(header)
print("    " + "-" * (5 * size))
for i, row in enumerate(table, 1):
    print(f"{i:>2} |" + "".join(f"{val:>5}" for val in row))

print(f"\nπ ≈ {math.pi:.10f}")
print(f"e ≈ {math.e:.10f}")
print(f"√2 ≈ {math.sqrt(2):.10f}")
```
